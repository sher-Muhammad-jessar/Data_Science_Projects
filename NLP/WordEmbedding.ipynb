```python
# libraries are tensorflow and keras
from tensorflow.keras.preprocessing.text import one_hot
```

```python
## sentecnes
sent=['the glass of milk',
      'the glass of juice',
      'the cup of tea',
      'i am good boy',
      'i am good devloper',
      'understand the meaning of words',
      'your videos are good'
]
```

```python
sent
```

```python
# one hot representation
voc_size=10000
onehot_repr=[one_hot(words, voc_size) for words in sent]
onehot_repr
```

#  word Embeddings Representation

```python
from tensorflow.keras.layers import Embedding
# pad sequences is used to make the same size of sentences
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length=8
#padding=pre means give zeros in front
embedded_docs=pad_sequences(onehot_repr, padding='pre', maxlen=sent_length)
print(embedded_docs)
```

```python
# to pass this in embedding layer we have to specify dimensions
dim=10 # features
```

```python
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```

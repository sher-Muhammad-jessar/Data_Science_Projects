```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
df.head()
```

```python
print(df["Churn"].value_counts())
```

```python
df["TotalCharges"]=pd.to_numeric(df["TotalCharges"], errors="coerce")
df.head(2)
```

```python
df.dropna(subset=["TotalCharges"], inplace=True)
```

```python
df
```

```python
drop_cols = ['customerID']
df.drop(columns=drop_cols, inplace=True)
```

```python
df
```

```python
# split features and target:
X=df.iloc[:,:-1]
Y=df["Churn"].map({'Yes':1,'No':0})
```

```python
X,Y
```

```python
# preprocessing : encoding and scaling 
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
numeric=X.select_dtypes(include=['int64','float64']).columns.tolist()
categorical=X.select_dtypes(include=['object']).columns.to_list()

preprocessor=ColumnTransformer([
    ('num',StandardScaler(),numeric),
    ('cat', OneHotEncoder(handle_unknown='ignore'),categorical)
])

clf=Pipeline( [('prep',preprocessor), ('rf',RandomForestClassifier(random_state=42))])

```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(X,Y, test_size=0.3, random_state=42, stratify=Y)
```

```python
# hyper parameter tunning 
from sklearn.model_selection import GridSearchCV

param = {
    'rf__n_estimators': [100, 200],
    'rf__max_depth': [None, 10, 20],
    'rf__max_features': ['sqrt', 'log2']
}


model=GridSearchCV(clf,param, cv=5 ,scoring='roc_auc', n_jobs=-1)
model.fit(X_train, Y_train)
```

```python
best=model.best_estimator_
best
```

```python
Y_pred=best.predict(X_test)
```

```python
Y_prob=best.predict_proba(X_test)[:,1]
```

```python
from sklearn.metrics import classification_report, roc_auc_score
print(classification_report(Y_test, Y_pred))
print("ROC-AUC:", roc_auc_score(Y_test, Y_prob))
```

```python
import shap

X_test_transformed = best.named_steps['prep'].transform(X_test)

explainer = shap.Explainer(best.named_steps['rf'], X_test_transformed)
shap_values = explainer(X_test_transformed, check_additivity=False)  # disable strict check
shap.summary_plot(shap_values, features=X_test_transformed, feature_names=explainer.feature_names)
```

```python
import joblib

# Save the best model
joblib.dump(best, "churn_model.pkl")
```
